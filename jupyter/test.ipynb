```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
c = 10  # Maximum value y approaches
d = 7   # Value of y when x = 0
k = 0.1 # Rate constant

# Logarithmic equation
def logarithmic_equation(x):
    return c - (c - d) * np.exp(-k * x)

# Generate x values
x_values = np.linspace(0, 50, 100)  # x from 0 to 50

# Calculate y values
y_values = logarithmic_equation(x_values)

# Plot the graph
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=r"$y = c - (c - d) \cdot e^{-k \cdot x}$", color="blue")
plt.axhline(y=c, color="red", linestyle="--", label=f"y = {c} (Asymptote)")
plt.axhline(y=d, color="green", linestyle="--", label=f"y = {d} (Initial Value)")
plt.title("Logarithmic Growth Curve")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Given data
x_data = np.array([0, 2000, 5000, 10000, 15000, 20000, 40000])  # Field values
y_data = np.array([10, 9, 8, 7, 6, 4, 4])                       # Sizes

# Define the equation
def equation(x, k):
    return -3 + 13 * np.exp(-k * x)

# Perform curve fitting
params, _ = curve_fit(equation, x_data, y_data, p0=[0.0001])

# Extract the fitted parameter
k_fit = params[0]
print(f"Fitted Parameter: k = {k_fit}")

# Generate fitted curve
x_fit = np.linspace(0, 40000, 500)
y_fit = equation(x_fit, k_fit)

# Plot the data and the fitted curve
plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_data, color="red", label="Data Points")
plt.plot(x_fit, y_fit, color="blue", label="Fitted Curve")
plt.title("Smooth Curve Fitting")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
